```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
```

```python
data = pd.read_csv('cancer.csv')
data.head()
```

```python
data.isnull().sum()
```

```python
data.describe()
```

```python
sns.heatmap(data.isnull(), annot=False)
```

```python
data.drop(['id', 'Unnamed: 32'], axis=1, inplace=True)
```

```python
data
```

```python
for column in data.columns:
    sns.histplot(data[column], kde=True)
    plt.title(column)
    plt.show()
    
```

```python
data["diagnosis"] = data["diagnosis"].map({"M": 1, "B": 0})
```

```python
data["diagnosis"].value_counts()
```

```python
x = data.drop("diagnosis", axis=1)
y = data["diagnosis"]
```

```python
scalar = StandardScaler()
x_scaled = scalar.fit_transform(x)
```

```python
X_train, X_test, y_train, y_test = train_test_split(x_scaled, y, test_size=0.2, random_state=42)
```

```python
def knn_model(x_trains, y_trains, x_tests, y_tests, k):
    from sklearn.neighbors import KNeighborsClassifier
    knn = KNeighborsClassifier(n_neighbors=k, metric="euclidean", weights="uniform", algorithm="auto")
    knn.fit(x_trains, y_trains)
    y_pred = knn.predict(x_tests)
    acc = accuracy_score(y_tests, y_pred)
    cm = confusion_matrix(y_tests, y_pred)
    print("Confusion Matrix:", cm)
    print("Accuracy:", acc)
    sns.heatmap(cm, annot=True, fmt="d")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.title("Confusion Matrix")
```

```python
knn_model(X_train, y_train, X_test, y_test, 13)
```
